```python
pd.options.display.float_format = '{:,.2f}'.format
def summary(df):
    print(f'data shape: {train.shape}')
    summ = pd.DataFrame(train.dtypes, columns=['data type'])
    summ['#missing'] = train.isnull().sum().values
    summ['%missing'] = train.isnull().sum().values / len(df) * 100
    summ['#unique'] = train.nunique().values
    desc = pd.DataFrame(df.describe(include='all').transpose())
    summ['min'] = desc['min'].values
    summ['max'] = desc['max'].values
    summ['average'] = desc['mean'].values
    summ['standard_deviation'] = desc['std'].values
    summ['first value'] = train.loc[0].values
    summ['second value'] = train.loc[1].values
    summ['third value'] = train.loc[2].values

    return summ
summary(train).style.background_gradient(cmap='YlOrBr')
```

```python
greater_than_100 = []
less_than_20 = []

# Iterate through columns
for column in train.columns:
    unique_values = train[column].unique()
    print(f"Total Unique value in {column}: {len(unique_values)}")

    # Check conditions and append to respective lists
    if len(unique_values) < 500:
        greater_than_100.append(column)
    elif len(unique_values) < 20:
        less_than_20.append(column)

# Print the lists
print("\nColumns with unique values greater than 500:", greater_than_100)
print("\nColumns with unique values less than 20:", less_than_20)
```
